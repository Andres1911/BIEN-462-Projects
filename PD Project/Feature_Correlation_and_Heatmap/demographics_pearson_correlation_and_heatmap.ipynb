```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data from CSV file
columns_to_use = ['Group', 'Gender', 'Age', 'Height (meters)', 'Weight (kg)']
data = pd.read_csv(r'C:\Users\alexa\OneDrive - McGill University\Documents\McGill\Year 4 U3\Semester 2\BIEN 462\Final Project\Dataset 1 - Adjusted Demographics Correlation\adjusted_demographics.csv', usecols=columns_to_use)
print(data.columns.tolist())
# Check if the data loaded correctly
print(data.head(160))

# Pearson Correlation
correlation_matrix = data.corr(method='pearson')

# Generate Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Pearson Correlation Heatmap between Features')
plt.show()
```
